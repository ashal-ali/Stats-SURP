Filtering all post 2019

```python
filtered_df = df[df['year'] < 2019]
filtered_df
```

Joining the lag data and extra variables

```python
lag1 = pd.read_csv("1yrlag.csv")
lag1.head()
```

```python
merged_df = pd.merge(filtered_df, lag1, on='schol_d', how='left')
merged_df.head()
```

```python
lag2 = pd.read_csv("extra_data.csv")
lag2.head()
```

```python
final_df = pd.merge(merged_df, lag2, on='schol_d', how='left')
final_df.head()
```

```python
sm = final_df[final_df["sm_prev"] >= 0]
sm
```
